Deleting files

OS modules provides 3 ways of deleting files:
    - os.unlink(path) deletes a file at the path you provide
    - os.rmdir(path) deletes a folder (folder must be empty) at the path you provide
    - shutil.rmtree(path) the MOST dangerous, as it will remove all files and folders contained in the path.
    
NOTE: All of these methods CAN NOT be reversed! Which means that if you make a mistake you won't be able to recover
the file. Instead we will use the send2trash module. A safer alternative that sends deleted files to the trash bin
instead of parmenent removal.

```python
for folder, sub_folders, files in os.walk('/Users/jose/Documents/GIT_Dyn/waf_rules'):
    print(f"Currently looking at: {folder}")
    print('\n')
    print("The sub-folders are: ")
    for sub_fold in sub_folders:
        print(f"\t Subfolder: {sub_fold}")
        
    print('\n')
    print('The files are: ')
    for f in files:
        print(f"\t File: {f}")
    print('\n')
    
```
