```python
import requests
import bs4
import lxml
```

```python
# We will use: https://es.wikipedia.org/wiki/Grace_Murray_Hopper
```

```python
res = requests.get('https://es.wikipedia.org/wiki/Grace_Murray_Hopper')
```

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
```

```python
soup
```

```python
soup.select('.toctext') # grab all toctext class elements
```
