```python
raw_email = email_data[0][1]
```

```python
raw_email_string = raw_email.decode('utf-8')
```

```python
raw_email_string
```

```python
import email # grab the actual message from the string
```

```python
email_message = email.message_from_string(raw_email_string)
```

```python
email_message
```

```python
for part in email_message.walk():
    
    if part.get_content_type() == 'text/plain': # you can also expect "text/html" (which might use a link inside of email)
        body = part.get_payload(decode=True)
        print(body)
```
